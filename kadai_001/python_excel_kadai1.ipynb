```python
import openpyxl
from datetime import datetime

workbook = openpyxl.Workbook()

sheet = workbook.active

sheet['B2'] = '請求書'

# 「B2 の下」に1行追加 = 3行目に追加
sheet.insert_rows(3, 1)

sheet['B4'] = '株式会社ABC'
sheet['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
sheet['B6'] = 'TEL:03-1234-5678   FAX:03-1234-5678'
sheet['B7'] = '担当者名:鈴木一郎 様'

# 「B7 の下」に2行追加 = 8,9行目が空行になる
sheet.insert_rows(8, 2)

# ヘッダーを B10〜E10 に書き込む
header_row = 10
headers = ['商品名', '数量', '単価', '金額']

for col, value in enumerate(headers, start=2):
    sheet.cell(row=header_row, column=col, value=value)

# データを B11〜 に書き込む
data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000],
]

for r, row_data in enumerate(data, start=header_row + 1):
    for c, value in enumerate(row_data, start=2):
        sheet.cell(row=r, column=c, value=value)




# 合計値
sheet['E13'] = '=SUM(E11:E12)'

labels = ['合計', '消費税', '税込合計']

for i, label in enumerate(labels, start=15):
    sheet.cell(row=i, column=2, value=label)

sheet['E15'] = 35000      # 合計
sheet['E16'] = 3500       # 消費税
sheet['E17'] = 38500      # 税込合計


# No.を挿入
sheet['F4'] = 'No.'
sheet['G4'] = '00001'

# 日付を挿入
sheet['F5'] = '日付'

today = datetime.today().date()

# 日付を挿入
sheet['G5'] = today

# 表示形式を YYYY/MM/DD に
sheet['G5'].number_format = 'yyyy/mm/dd'

today_str = datetime.today().strftime('%Y%m%d')
workbook.save(f'請求書_{today_str}.xlsx')

```
