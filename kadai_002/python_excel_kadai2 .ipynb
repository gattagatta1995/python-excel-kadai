```python
import pandas as pd

# データフレームの作成
df = pd.DataFrame({
    '日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
    '売上': [100, 200, 150, 300, 250],
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店'],
})

# 平均売上の計算　meanメソッド
sales_average = df['売上'].mean()
df['平均売上'] = [sales_average] * len(df)


# 業績ランクを決める関数「performance」を設定　条件分岐
def performance(level, sales_average):
    result = ''
    if level >= sales_average + 50:
        result = 'A'
    elif level >= sales_average:
        result = 'B'
    else:
        result = 'C'
    return result

# 「業績ランク」列を作成し、関数「performance」を適用して値を設定
df['業績ランク'] = df['売上'].apply(lambda x: performance(x, sales_average))

# エクセルファイルへの出力
with pd.ExcelWriter('業績.xlsx') as writer:
    df.to_excel(writer, sheet_name='Sheet1', index=False)
```
